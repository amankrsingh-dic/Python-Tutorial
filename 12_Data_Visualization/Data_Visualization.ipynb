# Data Visualization in Python

Visualizing data helps you understand and communicate insights. Python offers several libraries for creating a wide range of plots and charts.

---

## Choosing the Right Plot
- **Line Plot:** Trends over time or ordered data
- **Bar Chart:** Compare quantities across categories
- **Box Plot:** Show distribution and outliers
- **Scatter Plot:** Visualize relationships between two variables
- **Heatmap:** Show data density or correlation

---

## Line Plot with matplotlib
A line plot is useful for visualizing trends over intervals.

````python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', color='teal', linestyle='--')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Plot Example")
plt.grid(True)
plt.show()
````

## Bar Chart
Bar charts are great for comparing quantities across categories.

````python
plt.figure(figsize=(5, 3))
plt.bar(["A", "B", "C"], [5, 7, 3], color=['#4c72b0', '#55a868', '#c44e52'])
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Bar Chart Example")
plt.show()
````

## Seaborn: Statistical Data Visualization
Seaborn is a powerful Python library built on top of matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

To use seaborn, you may need to install it first:
```bash
pip install seaborn
```

Let's see some examples:

````python
import seaborn as sns
import numpy as np
import pandas as pd

# Create sample data
data = pd.DataFrame({
    "x": np.arange(1, 11),
    "y": np.random.randint(10, 100, 10),
    "category": ["A", "B"] * 5
})

# Seaborn Lineplot
plt.figure(figsize=(6, 4))
sns.lineplot(x="x", y="y", data=data, marker="o", hue="category", palette="Set2")
plt.title("Seaborn Line Plot by Category")
plt.show()

# Seaborn Barplot
plt.figure(figsize=(5, 3))
sns.barplot(x="category", y="y", data=data, ci=None, palette="pastel")
plt.title("Seaborn Bar Plot")
plt.show()

# Seaborn Boxplot
plt.figure(figsize=(5, 3))
sns.boxplot(x="category", y="y", data=data, palette="muted")
plt.title("Seaborn Box Plot")
plt.show()
````

Seaborn also supports many other plot types, such as scatter plots, heatmaps, and pair plots. It is especially useful for visualizing relationships in data and for statistical analysis.
